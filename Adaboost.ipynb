```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('绝对标签改.csv')
```

```python
n_features = 49
n_group = 141
n_train = 134
```

```python
datanew = data. iloc[1:, 9:]
dataT = datanew.T
X = dataT[dataT.columns[:n_features+1]]
y = dataT[n_features+2].astype('int')
```

```python
X.head()
```

```python
idx = []
idx.append(0)
j=3
for i in range(n_group-1):
    idx.append(j)
    j+=3
import random
random.shuffle (idx)
print(idx)
```

```python
X_train = pd.DataFrame(data=None,columns=list(range(1,n_features+2)))
y_train = pd.DataFrame(data=None,columns=[n_features+2])
print(X_train)
print(y_train)
```

```python
j=0
for i in range(n_train):
    X_train.loc[j] = X.iloc[idx[i]]
    j+=1
    X_train.loc[j] = X.iloc[idx[i]+1]
    j+=1
    X_train.loc[j] = X.iloc[idx[i]+2]
    j+=1
k=0
for i in range(n_train):
    y_train.loc[k] = y.iloc[idx[i]]
    k+=1
    y_train.loc[k] = y.iloc[idx[i]+1]
    k+=1
    y_train.loc[k] = y.iloc[idx[i]+2]
    k+=1
```

```python
X_test= pd.DataFrame(data=None,columns=list(range(1,n_features+2)))
j=0
for i in range(n_train,n_group):
    X_test.loc[j] = X.iloc[idx[i]]
    j+=1
    X_test.loc[j] = X.iloc[idx[i]+1]
    j+=1
    X_test.loc[j] = X.iloc[idx[i]+2]
    j+=1
    
y_test = pd.DataFrame(data=None,columns=[n_features+2])
k=0
for i in range(n_train,n_group):
    y_test.loc[k] = y.iloc[idx[i]]
    k+=1
    y_test.loc[k] = y.iloc[idx[i]+1]
    k+=1
    y_test.loc[k] = y.iloc[idx[i]+2]
    k+=1
```

```python
# 使用values属性将DataFrame转换为NumPy数组
array_2d = y_train.values

# 使用flatten()方法将二维数组展平为一维数组
y_train = array_2d.flatten()

# 现在，array_1d就是一个一维数组
print(y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix,accuracy_score, recall_score, f1_score, precision_score

abc = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=5)) # 初始化
abc.fit(X_train,y_train)     # 模型训练

#print("建立的弱分类器：",abc.estimators_)
#print("分类误差：",abc.estimator_errors_)
#print("分类器权重：",abc.estimator_weights_)
#print("迭代速率：",abc.learning_rate)
```

```python
y_pred = abc.predict(X_test) # 模型预测
print(y_pred)
print("\n---------- 模型评价 ----------")
cm = confusion_matrix(y_test, y_pred) # 混淆矩阵
df_cm = pd.DataFrame(cm)              # 构建DataFrame
print("ConfusionMatrix",df_cm)      
print('Accuracy score:', accuracy_score(y_test, y_pred))                       # 准确率
print('Recall:', recall_score(y_test, y_pred, average='weighted'))             # 召回率
print('F1-score:', f1_score(y_test, y_pred, average='weighted'))               # F1分数
print('Precision score:', precision_score(y_test, y_pred, average='weighted')) # 精确度
```
